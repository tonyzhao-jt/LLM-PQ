```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# grouping result
record = {
    'cluster 3': {
        'group=2': [49322.43213, 1.173157692], # latency and overhead
        'group=1': [49322.43213, 63.34078503],
        'shaqh': [53588.1, 5.784]
    },
    'cluster 6': {
        'group=2': [42658.35716, 2.620892525], # latency and overhead
        'group=1': [42412.92366, 19.14231706],
        'shaqh': [68064.05667, 7.702705383]
    },
    'cluster 4': {
        'group=2': [127310.9842, 8.147095203], # latency and overhead
        'group=1': [133693.0892, 126.2617514],
        'shaqh': [108140.8217, 2.005858898]
    },
    'cluster 10': {
        'group=2': [127606.2089, 5.812438965], # latency and overhead
        'group=1': [112720.3723, 104.7353451],
        'shaqh': [110881.828, 2.110968828]
    },
}
```

```python
def plot_record(record):
    clusters = list(record.keys())

    group1_values = [record[c]['group=1'][0] for c in clusters] 
    group2_values = [record[c]['group=2'][0] for c in clusters]
    shaqh_values = [record[c]['shaqh'][0] for c in clusters]
    g1_overhead = [record[c]['group=1'][1] for c in clusters]
    g2_overhead = [record[c]['group=2'][1] for c in clusters]
    shaqh_overhead = [record[c]['shaqh'][1] for c in clusters]

    bar_width = 0.2
    r1 = np.arange(len(clusters))
    r2 = [x + bar_width for x in r1]
    r3 = [x + bar_width for x in r2]

    fig, ax = plt.subplots()
    ax.bar(r1, group2_values, width=bar_width, label='group=2')
    ax.bar(r2, group1_values, width=bar_width, label='group=1')
    ax.bar(r3, shaqh_values, width=bar_width, label='shaqh')

    ax.set_xticks([r + bar_width for r in range(len(clusters))])
    ax.set_xticklabels(clusters)
    ax.set_xlabel('Cluster')
    ax.set_ylabel('Latency')
    ax.legend()

    ax2 = ax.twinx()
    ax2.plot(r1, g1_overhead, color='red', label='group=1 second value')
    ax2.plot(r2, g2_overhead, color='red', label='group=1 second value')
    ax2.plot(r3, shaqh_overhead, color='red', label='group=1 second value')
    ax2.set_ylabel('Second Value')
    ax2.legend(loc='upper right')

    plt.show()
```

```python
plot_record(record)
```
